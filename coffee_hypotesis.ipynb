```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
file_path = 'cafe1.csv'
coffee = pd.read_csv(file_path)
coffee
```

```python
#Hypothesis 0: The price per capsule for "Organic" coffees is significantly higher on average than non-organic coffees.

#Hypothesis 1: There is a significant difference in the average intensity level between different collections (e.g., "Collection Compostable," "Reviving Origins," "Special Reserve," "World Explorations," "Café Joyeux," and "Barista Creations").
```

```python
# Extractinh data for "Organic" and "Non-Organic" coffees
organic_coffees = coffee[coffee['Organic'] == 'Yes']
non_organic_coffees = coffee[coffee['Organic'] == 'No']

# Performing a t-test for the price per capsule
t_stat, p_value = stats.ttest_ind(organic_coffees['Price for capsule'], non_organic_coffees['Price for capsule'], equal_var=False)
print(t_stat, p_value )

alpha = 0.05

# Checking if the p-value is less than the significance level
if p_value < alpha:
    print("Hypothesis 0: Reject the null hypothesis")
    print("There is a significant difference in the price per capsule for Organic and Non-Organic coffees.")
else:
    print("Hypothesis 0: Fail to reject the null hypothesis")
    print("There is no significant difference in the price per capsule for Organic and Non-Organic coffees.")
```

```python
# hypothesis 0: there is no siginificant difference in intensity of coffee in different collections
# hypothesis 1: there is a significand difference in intensity of coffee in different collections.
```

```python
# Removing all NaN columng 'Intensite" column
coffee2 = coffee.dropna(subset=['Intensite'])

# Creating a dictionary to store intensity data for each collection
collection_intensity = {}
collections = coffee2['Collection'].unique()

for collection in collections:
    collection_intensity[collection] = coffee2[coffee2['Collection'] == collection]['Intensite']

# one-way ANOVA
f_stat, p_value = stats.f_oneway(*collection_intensity.values())
print(f_stat, p_value)


alpha = 0.05

# Checking if the p-value is less than the significance level
if p_value < alpha:
    print( "Reject the null hypothesis")
    print("There is a significant difference in the average intensity level between different collections.")
else:
    print("Fail to reject the null hypothesis")
    print("There is no significant difference in the average intensity level between different collections.")
```

```python
# hypothesis 0 There is no significant difference in the average price per capsule between Espresso and Lungo sizes of cups.
# hypothesis 1 There is a significant difference in the average price per capsule between Espresso and Lungo sizes of cups.
```

```python
# Extracting data for "Espresso" and "Lungo" sizes
espresso_coffees = coffee[coffee['Size of cup type1'] == 'Espresso']
lungo_coffees = coffee[coffee['Size of cup type1'] == 'Lungo']

# Perform a t-test for the price per capsule
t_stat, p_value = stats.ttest_ind(espresso_coffees['Price for capsule'], lungo_coffees['Price for capsule'], equal_var=False)
print(t_stat, p_value)

alpha = 0.05

# Checking if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is a significant difference in the average price per capsule between Espresso and Lungo sizes.")
else:
    print("Fail to reject the null hypothesis")
    print("There is no significant difference in the average price per capsule between Espresso and Lungo sizes.")
```
